Bahut important algorithm hai.

## AdaBoostClassifier kya hai?

**AdaBoost = Adaptive Boosting**

Ye ek **ensemble learning** algorithm hai.

Idea:

> Ek strong model banane ke liye kai weak models ko combine karo.

---

## Weak Learner kya hota hai?

Aisa model jo thoda sa hi sahi predict karta ho.

Example:

```text
Accuracy = 60%
```

AdaBoost kehta hai:

```text
Weak Model 1
Weak Model 2
Weak Model 3
Weak Model 4
...
```

In sab ko mila ke ek strong model banao.

---

## Real-Life Analogy

Maan lo exam ka answer puchha.

Ek student se puchha:

```text
60% chance correct
```

Fir 20 students se puchha:

```text
Majority answer
```

Accuracy improve ho sakti hai.

AdaBoost kuch aisa hi karta hai.

---

# Sabse Important Idea

Normal models:

```text
Sab observations ko equal importance
```

AdaBoost:

```text
Jo observations galat predict hui,
unko zyada importance do.
```

Isi liye naam:

```text
Adaptive Boosting
```

## Example

Training data:

| Hours | Pass |

### First Model

Predict karta hai:

```text
1 → 0 ✔
2 → 0 ✔
3 → 1 ✖
8 → 1 ✔
9 → 1 ✔
```

Galti:

```text
3 → 1
```

AdaBoost bolega:

> Is point ko zyada importance do.

---

### Second Model

Ab us difficult point par zyada focus karega.

---

### Third Model

Aur difficult points par focus karega.

---

### Final Prediction

Sab models milkar vote karenge.

---

# Python Example

```python
from sklearn.ensemble import AdaBoostClassifier

X = [
    [25, 50000],
    [30, 60000],
    [45, 80000],
    [35, 120000],
    [50, 150000]
]

y = [0, 0, 1, 1, 1]

model = AdaBoostClassifier(
    n_estimators=50,
    random_state=42
)

model.fit(X, y)

prediction = model.predict([[40, 90000]])

print(prediction)
```

# Interview Definition

**AdaBoostClassifier** is an ensemble boosting algorithm that combines multiple weak learners (usually shallow decision trees) into a strong classifier. Each new learner focuses more on the samples that previous learners misclassified, improving overall performance.

---

## Jab yaad rakhna ho

```text
Decision Tree
    ↓
Random Forest = Many trees independently
    ↓
AdaBoost = Many weak trees, each correcting previous mistakes
```

Ye line yaad rahi to AdaBoost ka core idea kabhi nahi bhoologe.


**Short answer:**
Aaj ke time mein agar mujhe bina zyada information ke ek choose karna ho, to **Random Forest** zyada safe default choice hoti hai.

Lekin "better" dataset par depend karta hai.

---

# Core Difference

## Random Forest = Bagging

Bahut saare trees independently train hote hain.

```text
Tree 1
Tree 2
Tree 3
Tree 4
...
↓
Voting
↓
Final Prediction
```

Har tree apna kaam karta hai.

---

## AdaBoost = Boosting

Trees ek ke baad ek train hote hain.

```text
Tree 1
↓
Mistakes identify
↓
Tree 2
↓
Mistakes identify
↓
Tree 3
...
```

Har naya tree purane tree ki galti sudharne ki koshish karta hai.

# GradientBoostinClassifier

Bahut badhiya. Ab tum **AdaBoost** ke baad **GradientBoostingClassifier** par aa gaye ho, jo boosting family ka aur powerful member hai.

# Ek Line Mein

**GradientBoostingClassifier**:

> Har naya tree pichhle model ki mistakes (errors/residuals) ko reduce karne ki koshish karta hai.

---

# AdaBoost vs Gradient Boosting

### AdaBoost

Focus karta hai:

```text
Galat predicted samples
```

Unko zyada weight de deta hai.

---

### Gradient Boosting

Focus karta hai:

```text
Prediction error kitna hai?
```

Aur us error ko minimize karne ki koshish karta hai.

Isliye naam:

```text
Gradient + Boosting
```

---

# Real-Life Analogy

Maan lo house price predict karni hai.

Actual price:

```text
10 lakh
```

### First Tree

Predict:

```text
8 lakh
```

Error:

```text
2 lakh
```

---

### Second Tree

Error ko learn karega:

```text
+1.5 lakh
```

---

### Third Tree

Bacha hua error seekhega:

```text
+0.4 lakh
```

---

### Fourth Tree

Aur correction karega:

```text
+0.1 lakh
```

Final:

```text
8 + 1.5 + 0.4 + 0.1
=
10 lakh
```

Yahi Gradient Boosting ka core idea hai.

---

# Visualization

### Random Forest

```text
Tree1
Tree2
Tree3
Tree4
↓
Vote
```

---

### Gradient Boosting

```text
Tree1
 ↓
Tree2 fixes Tree1
 ↓
Tree3 fixes Tree2
 ↓
Tree4 fixes Tree3
```

Sequential learning.

# Python Example

```python
from sklearn.ensemble import GradientBoostingClassifier

X = [
    [25, 50000],
    [30, 60000],
    [45, 80000],
    [35, 120000],
    [50, 150000]
]

y = [0, 0, 1, 1, 1]

model = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=0.1,
    random_state=42
)

model.fit(X, y)

prediction = model.predict([[40, 90000]])

print(prediction)
```

### Interview Answer

**GradientBoostingClassifier** is an ensemble boosting algorithm that builds decision trees sequentially. Each new tree is trained to reduce the errors made by the previous trees. Key parameters include `n_estimators`, `learning_rate`, and `max_depth`. It often achieves higher accuracy than Random Forest but usually requires more tuning and is more sensitive to overfitting.
