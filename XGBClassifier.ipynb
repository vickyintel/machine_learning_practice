# ROC Curve Formula Components

(True Positive Rate)

TPR=\frac{TP}{TP+FN}

Ye Recall bhi kehlata hai.

### FPR

(False Positive Rate)

FPR=\frac{FP}{FP+TN}

## Interview Memory Trick

```text
Accuracy  = Overall correctness

Precision = Positive bola to kitni baar sahi?

Recall    = Actual positives mein se kitne pakde?

F1 Score  = Precision + Recall balance

Confusion Matrix = TP, TN, FP, FN table

Classification Report = Sab kuch ek saath
```

Agar tum ML interviews ki tayari kar rahe ho, to **Accuracy → Precision → Recall → F1 → Confusion Matrix** ka flow achchhe se samajh lo. Ye bahut frequently poocha jata hai.
